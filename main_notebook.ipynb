```python
import pandas as pd
import numpy as np
import json

pd.set_option('max_colwidth', 400)
```

```python
contacts_df = pd.read_excel('Resources/contacts.xlsx')
crowd_df = pd.read_excel('Resources/crowdfunding.xlsx')
```

```python
# data prep -- the contacts_df needs to be properly parsed.
column_names = ['contact_id', 'first_name', 'last_name', 'email']

rows = []

for row in contacts_df[3:].iterrows():
    rows.append(row[1].iloc[0])
```

```python
rows[:5]
```

```python
json_rows = []

for row in rows:
    json_rows.append(json.loads(row))
```

```python
json_rows[:5]
```

```python
contacts_df = pd.DataFrame(json_rows)

contacts_df.head()
```

```python
names = []


for row in contacts_df['name']:
    fullname = row.split()
    fname, lname = fullname[0], fullname[1]
    names.append({'first_name': fname, 'last_name': lname})
```

```python
names = pd.DataFrame(names)
```
